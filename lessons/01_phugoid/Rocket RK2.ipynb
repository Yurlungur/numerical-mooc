# The Rocket Equations Solved using RK2

### import statements

```python
import numpy
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib import rcParams
rcParams['font.family'] = 'serif'
rcParams['font.size'] = 16
```

### Global variables

```python
g = 9.81 # m/s^2 acceleration due to gravity
rho = 1.091 # kg/m^3 density of air
r = 0.5 # radius of the rocket
A = numpy.pi*(r**2) # rocket radius
ve = 325 # m/s exhaust speed
CD = 0.15 # coefficient of drag
mp0 = 100 # kg
ms=50
burn_rate = lambda t: 20 if t < 5 else 0
h0=0.
v0=0.
```

### And the methods we need

```python
def f(u,t):
    """
    Returns the RHS of the rocket equations given a state vector u and a time t
    """
    return numpy.array([u[1],
                        (-(ms+u[2])*g + burn_rate(t)*ve - 0.5*rho*u[1]*abs(u[1])*A*CD)/(ms+u[2]),
                        -burn_rate(t)])
```

Notice now the intermediate step between $u_n$ and $u_{n+1}$.

```python
def rk2_step(u, f, t, dt):
    """Returns the solution at the next time-step using 2nd-order Runge-Kutta.
    
    Parameters
    ----------
    u : array of float
        solution at the previous time-step.
    f : function
        function to compute the right hand-side of the system of equation.
    t : float
        The current time of the simulation
    dt : float
        time-increment.
    
    Returns
    -------
    u_n_plus_1 : array of float
        solution at the next time step.
    """
    u_star = u + 0.5*dt*f(u,t)
    return u + dt*f(u_star,t+.5*dt)
```

To test convergence we will want to be able to run the integrator for several different values of dt. So we make a wrapper that runs the integration.

```python
def integrate_rocket(dt):
    """
    Integrates the rocket equations with step size in time dt. Stops when the rocket crashes.
    Assumes the rocket starts at height h=0 at rest. The rocket has 100 kg of fuel.
    """
    u = numpy.array([h0,v0,mp0])
    t = 0.
    u_solutions = u
    t_solutions = numpy.array([t])
    while u[0] >= 0.0:
        u = rk2_step(u,f,t,dt)
        t = t + dt
        u_solutions = numpy.vstack((u_solutions,u))
        t_solutions = numpy.hstack((t_solutions,t))
    return u_solutions,t_solutions
```

### Trial run

Now just to make sure everything is working, let's solve the system for the same dt we used in the homework.

```python
u_solutions,t_solutions = integrate_rocket(0.1)
h = u_solutions[...,0] # An easier way to extract h.
plt.plot(t_solutions,h)
plt.xlabel('time')
plt.ylabel('height')
plt.title("Plot for dt = 0.1")
```

### Convergence tests

Now we want to check for convergence, so let's generate a bunch of solutions for differnet dts. (We'll use powers of 2.)

```python
number_resolutions = 3
dt_values = numpy.array([(0.1**i)*0.1 for i in range(number_resolutions)])
print "We use the following dt values: {}".format(dt_values)
u_values = numpy.empty_like(dt_values, dtype=numpy.ndarray)
t_values = numpy.empty_like(dt_values, dtype=numpy.ndarray)
for i, dt in enumerate(dt_values):
    u_values[i],t_values[i] = integrate_rocket(dt)
    
# In the end, we will be interested in h, so we extract that information.
h_values = numpy.array([u_solution[...,...,0] for u_solution in u_values])
```

Now let's extract and plot h for several different resolutions.

```python
lines = [plt.plot(t_values[i],h_values[i]) for i in range(number_resolutions)]
plt.xlabel('time')
plt.ylabel('height')
plt.legend(['dt = {}'.format(dt) for dt in dt_values])
plt.title('Rocket solution for many resolutions')
```

To check convergence, let's do something clever. We'll find the difference between adjacent solutions and then we'll rescale all of the solutions by an appropriate factor. If the Runge-Kutta method is actually second order, all of these differences should line up.

```python
def get_difference(h1,h2,t1,factor):
    """
    Given two solutions to the rocket equations, h1 and h2, returns the difference at the times where they are 
    both evaluated.
    
    t1 is the time array for the less resolved solution. Returns an appropriate time array.
    
    Assumes that dt for h2 is (1/factor) times dt for h1.
    """
    h2_truncated = numpy.array(h2[::factor]) # makes the dt step the same
    # But there might be an off-by-one error. If so, fix this.
    if len(h1) > len(h2_truncated):
        h1_truncated = h1[:-1]
        t1_truncated = t1[:-1]
    elif len(h2_truncated) > len(h1):
        h2_truncated = h2_truncated[:-1]
        h1_truncated = h1
        t1_truncated = t1
    else:
        h1_truncated = h1
        t1_truncated = t1
    difference = h1_truncated-h2_truncated
    return difference,t1_truncated

def get_rescale_factor(dt1,dt2,order):
    """
    Given several values of dt: dt1,d2 for a system that converges at order, 
    such that the solutions are of the form
    h1 = h_true + alpha * dt1^order
    h2 = h_true + alpha * dt2^order
    
    calculates the factor that one should rescale (h1-h3) by for a self-covergence test.
    """
    return dt1**order - dt2**order
```

Now we're ready. Let's get the differences and rescale!

```python
differences = numpy.empty((2),dtype = numpy.ndarray)
truncated_times = numpy.empty((2),dtype = numpy.ndarray)
differences[0],truncated_times[0] = get_difference(h_values[0],h_values[2],t_values[0],100)
differences[1],truncated_times[1] = get_difference(h_values[1],h_values[2],t_values[1],10)
rescale_factors = numpy.array([get_rescale_factor(dt_values[0],dt_values[2],2),
                               get_rescale_factor(dt_values[1],dt_values[2],2)])
differences= [differences[i]/rescale_factors[i] for i in range(len(differences))]
print rescale_factors
```
